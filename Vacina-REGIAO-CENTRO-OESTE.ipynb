```python
import requests 
import pandas as pd
import time
from datetime import datetime
import csv


import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


# Leitura do arquivo de VaciNA.csv selecionando alguns
# campos para pesquisa:
 
df = pd.read_csv("BD_Amostra-Alguns-26-05-2021.csv", sep=";")
usecols = ['vacina_fabricante_nome',
                 'paciente_idade',
                 'paciente_id',
                 'vacina_dataAplicacao',
                 'paciente_endereco_uf',
                 'vacina_descricao_dose']
```

```python
# A porcentagem da Base de Vacinas é 1/1000

FATOR_AMOSTRA = 0.001 
```

```python
# Total  de pessoas vacinadas 
# na Região Centro-Oeste:

# (1) Mato Grosso - MT

Mato = df.loc[(df.paciente_endereco_uf == 'MT')]

print(len(Mato))

Mato.hist(bins=10)
```

```python

# (2) Mato Grosso do Sil - MS

Mato_Sul = df.loc[(df.paciente_endereco_uf == 'MS')]

Mato_Sul.hist(bins=10)
```

```python

# (3) Goiás - GO

Goias = df.loc[(df.paciente_endereco_uf == 'GO')]

Goias.hist(bins=10)
```

```python

# (4) Distrito Federal - DF

DF = df.loc[(df.paciente_endereco_uf == 'DF')]

print(len(DF))

DF.hist(bins=10)
```

```python
# Calculando o total de vacinados da Região Centro-Oeste
#

print ("Somando total de vacinados .......")


soma_1 = df.loc[df['paciente_endereco_uf'].
                isin(['MT', 'MS', 'GO', 'DF'])]

print("A soma total da Regiao CENTRO-OESTE --------> ", len(soma_1)/FATOR_AMOSTRA)                             
                                         
soma_1.hist(bins=10)
```

```python
# O calculo de porcentagem total dos vacinados na REgiao Centro-Oeste sera o
# valor dos vacinados divido pela populacao dos Brasileiros e o FATOR AMOSTRA:

Porcentagem = len(soma_1)/(211755692)

print('A porcentagem REAL brasileiros vacinados Regiao CENTRO_OESTE é: ', Porcentagem/FATOR_AMOSTRA)
              
```
