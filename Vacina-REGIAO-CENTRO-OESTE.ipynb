```python
import requests 
import pandas as pd
import time
from datetime import datetime
import csv


import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


# Leitura do arquivo de VaciNA.csv selecionando alguns
# campos para pesquisa:

df = pd.read_csv("vacinas_2021-04-03-1708.csv", sep=";", 
                 usecols=['vacina_descricao_dose', 'paciente_idade',
                          'vacina_categoria_nome',
                          'vacina_fabricante_nome', 'paciente_endereco_uf'])
```

```python
# Total  de pessoas vacinadas 
# na Região Centro-Oeste:

# (1) Mato Grosso - MT

Mato = df.loc[(df.paciente_endereco_uf == 'MT')]

print(len(Mato))

Mato.hist(bins=100)
```

```python

# (2) Mato Grosso do Sil - MS

Mato_Sul = df.loc[(df.paciente_endereco_uf == 'MS')]

Mato_Sul.hist(bins=100)
```

```python

# (3) Goiás - GO

Goias = df.loc[(df.paciente_endereco_uf == 'GO')]

Goias.hist(bins=100)
```

```python

# (4) Distrito Federal - DF

DF = df.loc[(df.paciente_endereco_uf == 'DF')]

print(len(DF))

DF.hist(bins=100)
```

```python
# Calculando o total de vacinados da Região Centro-Oeste
#


print ("Somando total de vacinados .......")


soma_1 = df.loc[df['paciente_endereco_uf'].
                isin(['MT', 'MS', 'GO', 'DF'])]


print("A soma total da Regiao CENTRO-OESTE --------> ", len(soma_1))                             
                                     
    
soma_1.hist(bins=100)
```

```python
Porcentagem = len(soma_1)/(212000)

print('A porcentagem brasileiros vacinadas na Regiao SUL é: ', Porcentagem)
              
```
