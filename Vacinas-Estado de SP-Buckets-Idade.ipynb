```python
# DATA ; 26/5/2021

import csv
import time
import requests 
import pandas as pd
import numpy  as np
from datetime import datetime

# Importar bibliotecas para construir gráficos:


import seaborn           as sns
import matplotlib.pyplot as plt
sns.set()


# Leitura do arquivo de Vacinacao completo selecionando alguns
# campos para pesquisa:

df = pd.read_csv("BD_Amostra-Alguns-26-05-2021.csv", sep=";", 
                 usecols = ['vacina_fabricante_nome',
                            'paciente_idade',
                            'paciente_id',
                            'vacina_dataAplicacao',
                            'paciente_endereco_uf',
                            'vacina_descricao_dose'])

                              
```

```python
# Fator amostra faz o calculo na base real:

FATOR_AMOSTRA = 0.001 # 1/1000
```

```python
# Tamanho do Arquivo atualizado, com o Programa: "Rodar-PRINCIPAL-salvar-Vacinas")

df.shape
```

# Total  de pessoas vacinadas 
# no Estado de SP 

```python
#Selecionando o campo cujo endereço de São Paulo - SP

df.loc[df['paciente_endereco_uf'] == 'SP'].hist(bins=20, 
                figsize = (10,10))

# População do Estado de São Paulo (SP) de acordo com CNN Brasil:46.289.333

print('Total de vacinados em São Paulo na AMOSTRA- SP ==>', 
      len(df.loc[df.paciente_endereco_uf == 'SP']))


Sao_Paulo = len(df.loc[df['paciente_endereco_uf'] == 'SP'])/46289333

Vacinados_sp = (Sao_Paulo)/FATOR_AMOSTRA

print('Porcentagem Real dos vacinados em SP:', Vacinados_sp)
```
